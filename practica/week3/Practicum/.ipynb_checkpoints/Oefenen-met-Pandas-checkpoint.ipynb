## Opdracht 3: samenstellen

Stel dat Nutella zou bestaan uit 50% cacao, 20% suiker, 5% mais en 25% kokosolie.

Verder is gegeven:

- voor deze grondstoffen is de prijs per kilogram
- een pot bevat 800 gram Nutella

Wat is dan de prijs voor een pot Nutella in elk jaar?

```python
# Voeg een kolom 'Nutella' toe aan het DataFrame:

# Bereken de prijs voor een pot van 800 gram over de jaren heen:

# Als we uitgaan van prijs per kg, willen we dus elk ingrediënt (het percentage daarvan in de 800gr) keer 0.8 doen om 
# de benodigde hoeveelheid te krijgen. Voor cacao is dat dus: 0.800g * 0.5 = 0.400g. Dit kunnen we vermendigvuldigen met
# de prijs per kilo, voor elk jaar. We doen dit voor alle ingrediënten, en voegen dit toe in een kolom 'Nutella':


df['Nutella'] = (df['Cocoa']*(0.8*0.5) + df['Sugar']*(0.2*0.8) + df['Maize']*(0.05*0.8) + df['Coconut Oil']*(0.25*0.8))

# Bereken de prijs voor een pot van 800 gram over de jaren heen:
```
